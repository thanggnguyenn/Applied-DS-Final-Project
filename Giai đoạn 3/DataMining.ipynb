```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import xgboost as xgb
import catboost

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
from lightgbm import LGBMClassifier
```

```python
data = pd.read_csv("feature_extracted_data.csv",index_col="CLIENTNUM")
data.head()
```

```python
data.shape
```

```python
data.dtypes
```

```python
target = "Attrition_Flag"
X = data.drop(target,axis=1)
y = data[target]
```

```python
X_train, X_test,y_train, y_test = train_test_split(X,y,random_state=42)
```

## Huấn luyện dữ liệu

### Random Forest

```python
'''
  n_jobs = -1 --> all CPUs are used
  class_weight = balanced --> The “balanced” mode uses the values of y to automatically adjust weights inversely proportional to
  class frequencies in the input data as n_samples / (n_classes * np.bincount(y))
  max_depth = 5 --> the maximum depth of the tree
'''
forest = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=30)

forest.fit(X_train,y_train)
```

```python
forest_pred = forest.predict(X_test)
```

```python
print(classification_report(forest_pred,y_test,target_names=["Existing Customer","Attrited Customer"]))
```

```python
forest_score = pd.DataFrame({"Precision": [precision_score(y_test,forest_pred)],\
                            "Recall":[recall_score(y_test,forest_pred)],
                             "ROC_AUC":[roc_auc_score(y_test,forest_pred)]})
forest_score
```

### XGBoost

```python
xgb_cl = xgb.XGBClassifier(learning_rate=0.5,\
                            max_depth=30)
xgb_cl.fit(X_train,y_train)
```

```python
xgboost_pred = xgb_cl.predict(X_test)

print(classification_report(y_test,xgboost_pred))
```

```python
xgboost_score = pd.DataFrame({"Precision": [precision_score(y_test,xgboost_pred)],\
                            "Recall":[recall_score(y_test,xgboost_pred)],
                             "ROC_AUC":[roc_auc_score(y_test,xgboost_pred)]})
xgboost_score
```

### CatBoost

```python
from catboost import CatBoostClassifier

# Initialize CatBoostClassifier
catboost_model = CatBoostClassifier(\ # iterations = 500,\
                                    learning_rate=0.5,
                                    depth=10)
catboost_model.fit(X_train,y_train)
print("Finished!")
```

```python
catboost_pred = catboost_model.predict(X_test)

print(classification_report(catboost_pred,y_test))
```

```python
catboost_score = pd.DataFrame({"Precision": [precision_score(y_test,catboost_pred)],\
                            "Recall":[recall_score(y_test,catboost_pred)],
                             "ROC_AUC":[roc_auc_score(y_test,catboost_pred)]})
catboost_score
```

### LightGBM

```python
lgbm_model = LGBMClassifier(max_depth=30,\
                            learning_rate=0.5,\
                            class_weight='balanced')

lgbm_model.fit(X_train,y_train)
```

```python
lgbm_pred = lgbm_model.predict(X_test)

print(classification_report(y_test,lgbm_pred))
```

```python
lgbm_score = pd.DataFrame({"Precision": [precision_score(y_test,lgbm_pred)],\
                            "Recall":[recall_score(y_test,lgbm_pred)],
                             "ROC_AUC":[roc_auc_score(y_test,lgbm_pred)]})
lgbm_score
```
